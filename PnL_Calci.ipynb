```python
import tabula
file_path = 'E:/company tests/silicon-2024-Yashas-Kishore-main/pnl_calculator/input/Investments-1.pdf'

# convert PDF into CSV
tabula.convert_into(file_path, "output.csv", output_format="csv", pages='all')
```

```python
from PyPDF2 import PdfReader

reader = PdfReader(file_path)
#assuming the all column header is 2 words 
# Extract the first line of text from the first page
first_page = reader.pages[0]
first_line = (first_page.extract_text().splitlines()[0]).split()
header = [f"{first_line[i]} {first_line[i+1]}" for i in range(0, len(first_line), 2)]
```

```python
import pandas as pd
data = pd.read_csv("E:/company tests/silicon-2024-Yashas-Kishore-main/pnl_calculator/output.csv", header=None)
# Manually assign column names
data.columns = header
data
```

```python
# Filter out the rows where 'Date Sold' is NaN (i.e., stocks that have not been sold)
sold_stocks = data.dropna(axis=0, how='any', subset=['Date Sold'])
sold_stocks
```

```python
# Calculate the profit/loss for each sold stock
sold_stocks['Profit/Loss'] = (sold_stocks['Sale Price/Share'] - sold_stocks['Purchase Price/Share']) * sold_stocks['Shares Purchased']

# Group by 'Stock Symbol' and sum up the profit/loss
grouped_profit_loss = sold_stocks.groupby('Stock Symbol')['Profit/Loss'].sum().reset_index()

# Rename columns
grouped_profit_loss.columns = ['Stock Symbol', 'Net Profit/Loss (PnL)']

# Print the results in the specified format
for index, row in grouped_profit_loss.iterrows():
    print(f"{row['Stock Symbol']},{row['Net Profit/Loss (PnL)']:.1f}")
```
